- No missing values in Size column so, we are good to go

- Verify the number of values  and three different kind of values

```python
# find the values  having M in them
df['Size'].loc[df['Size'].str.contains('M')].value_counts().sum()
```

```python
# find the values  having k in them
df['Size'].loc[df['Size'].str.contains('k')].value_counts().sum()
```

```python
# find the values  having "varies in them" in them
df['Size'].loc[df['Size'].str.contains('Varies with device')].value_counts().sum()
```

```python
# convert the whole column into the bythes
# let's define the function 
def convert_size (size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', "")) * 1024
        elif 'M' in size:
            return float(size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
# lets apply this function 
df['Size'] = df['Size'] . apply(convert_size)
```

```python
df['Size']
```

```python
# rename the column 
df.rename(columns={'Size':"Size_in_bytes"}, inplace=True)
```

- install column have no null and missing values so we are good to go , the only + sign we are going to replace it and also remove the , as well.
- So, lets deal with Install column.

### 1- Remove + sign 

```python
df['Installs'] = df['Installs'].apply(lambda x : x. replace('+', "") if '+' in str (x) else x)
```

```python
df['Installs'].value_counts()
```

### 2- Remove comma,

```python
df['Installs'] = df['Installs'].apply(lambda x : x. replace(',', "") if ',' in str (x) else x)
```

```python
df['Installs'].value_counts()
```

### 3-Convert the column into the integer

```python
df['Installs'] = df['Installs'].fillna(0)
```

```python
df['Installs'].value_counts()
```

# Missing values inside the DataFrame

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# find missing values percentage in the data 
missing_percentage = df.isnull().mean().round(2) * 100
print(missing_percentage)
```

# Plot missing values
## Bar plot:


```python
# Plot missing values
import matplotlib.pyplot as plt

missing_percentage.plot(kind='bar')
plt.title('Missing Values Percentage')  
plt.xlabel('Features')
plt.ylabel('Percent')
plt.show()
```

## Vertical bar plot

```python
missing_percentage.plot(kind='bar', figsize=(10,8), rot=0)
plt.grid(axis='y', alpha=0.75)
plt.title('Missing Values Percentage by Feature')
plt.ylabel('Percent Missing')
plt.show()
```

## Horizontal bar plot

```python
missing_percentage.plot(kind='barh', figsize=(10,8))
plt.grid(axis='x', alpha=0.75)
plt.title('Missing Values Percentage by Feature')
plt.xlabel('Percent Missing')  
plt.show()
```

## Pie chart

```python
# Plot pie chart

features = missing_percentage.index
sizes = missing_percentage

explode = [0] * len(features) 
explode[1] = 0.1

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=features, autopct='%1.1f%%',
       shadow=True, startangle=90)

ax1.axis('equal')  
plt.title('Missing Values Percentage Breakdown')
plt.show()
```

## Heatmap

```python
# Convert to DataFrame
missing_pct = missing_pct.to_frame()  

# Plot heatmap
plt.figure(figsize=(16,6))
sns.heatmap(missing_pct, cmap='viridis', center=0, linewidths=0.5) 

# Add labels and title
plt.yticks(rotation=0)  
plt.xlabel('Features', fontsize=14)
plt.ylabel('Percent Missing', fontsize=14)
plt.title('Missing Value Heatmap', fontsize=16)

plt.show()
```
